```python
type_shift = ''
#type_shift = '_shift' 

# import build-in
import numpy as np
import os
import random
from shutil import copyfile, copy2 , move, rmtree
from natsort import natsorted
import glob
from tqdm import tqdm

# import local
import utility_classification as ut_cla
import labels_create_ENSO as labels
```

```python
path = '/media/nicola/Nicola_data/class/canesm2/regions_concat'+type_shift+'/'

features = ["nino3W.nc","nino3E.nc","nino4W.nc","nino4E.nc","nino12.nc"]
file_id = "ens"

dict_data_struc=ut_cla.create_data_struct(path,features,file_id=file_id)
dict_data_struc
```

```python
#process all data
result = ut_cla.preprocess_from_dict(dict_data_struc, path, verbose=False)
```

```python
# Concatenate 
def concat_features(result, keys, axis=0):
    list_= [result[key] for key in keys]
    return np.concatenate(list_,axis=axis)
    
keys=list(range(0,50)) # key to concatenate
concat_N3W= concat_features(result, keys)

keys=list(range(50,100)) # key to concatenate
concat_N3E= concat_features(result, keys)

keys=list(range(100,150)) # key to concatenate
concat_N4W= concat_features(result, keys)

keys=list(range(150,200)) # key to concatenate
concat_N4E= concat_features(result, keys)

keys=list(range(200,250)) # key to concatenate
concat_N12= concat_features(result, keys)
##

input_train=np.concatenate([concat_N3W,concat_N3E,concat_N4W,concat_N4E,concat_N12],axis=1)
```

```python
import joblib

filename = 'ensemble_classifier_final.sav'
clf =joblib.load(filename)

prediction2=clf.predict(input_train)

with open('classified_model_output/canesm2'+type_shift+'.txt', 'w') as filehandle:  
    for listitem in prediction2:
        filehandle.write('%s\n' % listitem)
```
