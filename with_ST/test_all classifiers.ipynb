```python
#test all the classifiers available and print out scores

# define classifiers name
names = ["NearestNeighbors", "LinearSVM", "RBFSVM","DecisionTree", "RandomForest", "NeuralNet", "AdaBoost",
         "NaiveBayes", "QDA"]

# define classifier setting
classifiers = [
    KNeighborsClassifier(3),
    SVC(kernel="linear", C=0.025),
    SVC(gamma=2, C=1),
    #GaussianProcessClassifier(1.0 * RBF(1.0)),
    DecisionTreeClassifier(max_depth=5),
    RandomForestClassifier(max_depth=5, n_estimators=10, max_features=1),
    MLPClassifier(alpha=1, max_iter=400),
    AdaBoostClassifier(),
    GaussianNB(),
    QuadraticDiscriminantAnalysis()]


for name, clf in zip(names, classifiers):
    print(name)
    #ax = plt.subplot(1, len(classifiers) + 1, i)
    clf.fit(input_train, labels_train)
    score = clf.score(input_test, labels_test)
    pred=clf.predict(input_test)
    print('cross val',cross_val_score(clf, input_train, labels_train, cv=5).mean())
    print('accuracy score', accuracy_score(labels_test, pred))
    print('precision score CP/EL/LN/NE/ST', precision_score(labels_test, pred,labels=['CP','EL','LN','NE','ST'], average=None))
```
